```python
import  numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x=np.arange(0,5,0.1)
y=np.sin(x)

fig, ax=plt.subplots()
ax.plot(x,y,)
ax.set_title("Sample Graph", fontdict={'fontname':'Comic Sans MS','fontsize':20})
ax.set_ylabel("Y-axis",fontdict={'fontname':'Times New Roman'})
ax.set_xlabel("X-axis",fontdict={'fontname':'Times New Roman'})
```
